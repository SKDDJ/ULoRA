```python
import os
import torch
# 
# checkpoint_path = '/root/shiym_proj/Sara/vit-finetune/output/INIT-svd-768-1-base-aircraft-R-768-LR-5e-2-SCHEDULER-linear/version_0/checkpoints/best-step-step=240-val_acc=0.4527.ckpt'
checkpoint_path = '/root/shiym_proj/Sara/vit-finetune/output/INIT-svd-768-1-base-flowers102-R-768-LR-5e-2-SCHEDULER-linear/version_0/checkpoints/best-step-step=24-val_acc=0.9902.ckpt'

save_dir = "/root/shiym_proj/Sara/vit-finetune/vector_z_flowers_real"
os.makedirs(save_dir, exist_ok=True)
count_q = 0  
count_v = 0  

# 
checkpoint = torch.load(checkpoint_path, map_location=torch.device('cpu'))

# 
# print(checkpoint.keys())
print(checkpoint['state_dict'].keys())

for k, v in checkpoint['state_dict'].items():
    if 'vector_z' in k:
        vector_z = v
        if 'query' in k:
            query = vector_z
            print('Shape:', query.shape)
            print('Values:', query)
            count_q += 1
            save_path = os.path.join(save_dir, f"layer_{count_q}_vector_z_query.pt")
            torch.save(vector_z, save_path)
            print(f"Saved vector_z_key of layer {count_q} to {save_path}")
        else:
            count_v += 1
            save_path = os.path.join(save_dir, f"layer_{count_v}_vector_z_value.pt")
            torch.save(vector_z, save_path)
            print(f"Saved vector_z_value of layer {count_v} to {save_path}")
        

# #  vector_z 
# if 'vector_z' in checkpoint['state_dict']:
#     vector_z = checkpoint['state_dict']['vector_z']
#     if 'query' in checkpoint['state_dict']:
#         query = vector_z
#         print('Shape:', query.shape)
#         print('Values:', query)
#         count_q += 1
#         save_path = os.path.join(save_dir, f"layer_{count_q}_vector_z_query.pt")
#         torch.save(vector_z, save_path)
#         print(f"Saved vector_z_key of layer {count_q} to {save_path}")
#     else:
#         count_v += 1
#         save_path = os.path.join(save_dir, f"layer_{count_v}_vector_z_value.pt")
#         torch.save(vector_z, save_path)
#         print(f"Saved vector_z_value of layer {count_v} to {save_path}")
# else:
    # print('vector_z not found in the checkpoint')
```
